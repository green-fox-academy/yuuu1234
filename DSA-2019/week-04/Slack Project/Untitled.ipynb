```python
all_user=[]
m_user1=list(mentions.user1)
m_user2=list(mentions.user2)
r_user1=list(reactions.user1)
r_user2=list(reactions.user2)
all_user.extend(m_user1)
all_user.extend(m_user2)
all_user.extend(r_user1)
all_user.extend(r_user2)
all_user=list(set(all_user))
```

# relationship count between 2 users

```python
relationship_count_matrix=[]
for i in all_user:
    row=[]
    for j in all_user:
        count=count_relationship(i,j)
        row.append(count)
    relationship_count_matrix.append(row)
```

```python
np.asarray(relationship_count_matrix)
```

```python
df=pd.DataFrame(np.asarray(relationship_count_matrix),columns=all_user,index=all_user)
df
```

```python
plt.pyplot.figure(figsize=(20, 20))
sns.heatmap(df,linewidths=.5,cmap="YlGnBu",annot=True)
plt.show()
```

```python
def count_relationship(user1,user2):
    count=0
    for i in range(len(m_user1)):
        user_pair=[m_user1[i],m_user2[i]]
        if user1 in user_pair and user2 in user_pair and user1!=user2:
            count+=1
    for j in range(len(r_user1)):
        user_pair=[r_user1[i],r_user2[i]]
        if user1 in user_pair and user2 in user_pair and user1!=user2:
            count+=1
    return count
            
```
