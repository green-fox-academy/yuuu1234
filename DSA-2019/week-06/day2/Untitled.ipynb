```python
#lap_top_name=laptops_info.find_all('a',{'class':'name browsinglink'})
def collect_laptop_info(content):
    laptops=[]
    for lp in content:
        laptop={}
        laptop_name=lp.find('a',attrs={'class':'name browsinglink'}).text
        inner_price=lp.find('div',attrs={'class':'priceInner'})
        current_price=inner_price.find('span',attrs={'class':'c2'}).text
        previous_price=none_type_check(inner_price.find('span',attrs={'class':'np2'}))
        promotion=none_type_check(inner_price.find('span', attrs={'class':'np'}))
        exclude_VAT=inner_price.find('b').text
        stock=lp.select('span[class*="avlVal avl"]')[0].text
        try:
            rating=lp.find('div', attrs={'class':'star-rating-wrapper'})["data-rating"]
        except KeyError:
            rating=""
        #print(rating)
        #stock=lp.find('span', attrs={'class':'avlVal avl0 variant'}).text
        laptop['laptop_name']=laptop_name
        laptop['current_price']=current_price
        laptop['previous_price']=previous_price
        laptop['promotion']=promotion
        laptop['exclude_VAT']=exclude_VAT
        laptop['stock']=stock
        laptop['rating']=rating
        laptops.append(laptop)
        #break
    return laptops
#     print(laptop_name)
#     print(current_price)
#     print(previous_price)
#     print(promotion)
#     print(exclude_VAT)
#     print(stock)
```

```python
def none_type_check(value):
    result=""
    try: 
        result=value.text
    except AttributeError:
        pass
    return result
```
