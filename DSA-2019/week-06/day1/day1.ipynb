```python
#What are the ten most recently updated apps?
import re
#df2.sort_values(["Last Updated"
#pd.to_datetime(df2[0,'Last Updated']).dt.time
def time_convertion(value):
    time=""
    #time= np.where(re.match(r"\w+\s\d{1,2}\,\s\d{4}",value) is not None,re.match(r"\w+\s\d{1,2}\,\s\d{4}",value).group(),"false")
    try:
        time=pd.to_datetime(re.match(r"\w+\s\d{1,2}\,\s\d{4}",value).group())
    except:
        time="NaN"
    return time
        
df2["Last updated time"]=df2["Last Updated"].apply(time_convertion)
df2.sort_values(["Last updated time"],ascending=False)
```

```python
#Find apps, that is in the top 25% of apps based on ratings and were updated in the last two months of dataset.
df2["Last updated month"]=df2["Last updated time"].dt.month
df2["Last updated year"]=df2["Last updated time"].dt.year
#df2[df2["Rating"]<=df2.describe().loc["25%","Rating"] and 
```

```python
#Which app not to download? - Find an app which takes the most space on your phone, but is rated poorly (define poorly)?
def size_normalization(value):
    if "M" in value.upper():
        value=value
    elif"K" in value.upper():
        value=str(round(float(value[:-1])/1000,2))+"M"
    else:
        value="0M"
    return value
df2["Normalized size"]=df2["Size"].apply(size_normalization)
```

```python
def convert_m_to_million(value):
    try:
        float(value)
    except ValueError:
        value = str(float(value[:-1]) * 1000000)
    return value
df2["Reviews"]=df2["Reviews"].apply(convert_m_to_million)
```

```python
 df2.sort_values(['Reviews'])
```

```python
#reate a reviews reliability of reviews index, which takes into account, 
#how many reviews does the app have and how high the rating is. What metric do you take? Create as a new field.
df2["outlier"]=0
review_mean=df2["Reviews"].mean()
review_std=df2["Reviews"].std()
rating_mean=df2["Rating"].mean()
rating_std=df2["Rating"].std()
df2.loc[(df2["Rating"]<rating_mean-3*rating_std) | (df2["Rating"]>rating_mean+3*rating_std) ,"outlier"]=1
df2.loc[df2["Reviews"]<df2.describe().loc["25%","Reviews"],"outlier"]=1
```
